```python
import plotly.express as px
import os
import json

df["neighborhood"] = df["subtitle"].str.extract(r"^(.*?),\s*Madrid")

df["district"] = df["district"].replace({
    "Chamartín": "Chamartin",
    "Tetuán": "Tetuan",
    "Chamberí": "Chamberi",
    "Vicálvaro": "Vicalvaro"
})

price_mean = df.groupby("district")["buy_price_by_area"].mean().reset_index(name="price_avg")
price_median = df.groupby("district")["buy_price_by_area"].median().reset_index(name="price_med")
n_ads = df.groupby("district").size().reset_index(name="ads_count")

gdf = gpd.read_file("madrid-districts.geojson")
gdf = gdf.drop(columns=["created_at", "updated_at"], errors="ignore")

merged = gdf.merge(price_mean, how="left", left_on="name", right_on="district")
merged = merged.merge(price_median, how="left", left_on="name", right_on="district")
merged = merged.merge(n_ads, how="left", left_on="name", right_on="district")

geojson = json.loads(merged.to_json())

fig = px.choropleth_mapbox(
    merged,
    geojson=geojson,
    locations="name",
    featureidkey="properties.name",
    color="price_avg",
    color_continuous_scale="RdBu_r",
    mapbox_style="carto-positron",
    zoom=9.5,
    center={"lat": 40.42, "lon": -3.70},
    opacity=0.65,
    labels={"price_avg": "€/m²"},
    hover_name="name",
    hover_data={
        "price_avg": True,
        "price_med": True,
        "ads_count": True
    }
)

fig.update_layout(
    autosize=True,
    margin={"r":0, "t":0, "l":0, "b":0}
)

fig.show()
os.makedirs("site/assets", exist_ok=True)

fig.write_html("site/assets/madrid_map.html", full_html=True, include_plotlyjs="cdn")

# Injecter le style responsive
with open("site/assets/madrid_map.html", "r", encoding="utf-8") as f:
    html = f.read()

injection = """
<style>
  html, body {
    margin: 0;
    padding: 0;
    height: 100%;
  }

  .plotly-graph-div {
    height: 90vh !important;
    width: 100vw !important;
  }
</style>
"""

html = html.replace("</head>", injection + "\n</head>")

with open("site/assets/madrid_map.html", "w", encoding="utf-8") as f:
    f.write(html)
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import os

# Load the dataset
df = pd.read_csv("houses_Madrid.csv")

# Keep only relevant columns
df_clean = df[[
    'sq_mt_built',
    'buy_price',
    'neighborhood_id'
]].copy()

# Remove rows with missing values
df_clean = df_clean.dropna(subset=['sq_mt_built', 'buy_price', 'neighborhood_id'])

# Rename columns for clarity
df_clean.columns = ['Surface (m²)', 'Price (€)', 'Neighborhood']

# Group by surface (rounded to nearest 10) to compute average price
df_clean['Surface Group'] = (df_clean['Surface (m²)'] // 10) * 10
avg_price_by_surface = df_clean.groupby('Surface Group')['Price (€)'].mean().reset_index()

# Create base scatter plot WITHOUT using px (for layering control)
scatter_trace = go.Scatter(
    x=df_clean['Surface (m²)'],
    y=df_clean['Price (€)'],
    mode='markers',
    marker=dict(
        color=df_clean['Neighborhood'].astype('category').cat.codes,
        colorscale='Viridis',
        showscale=False,
        opacity=0.8,
        size=5
    ),
    text=df_clean['Neighborhood'],
    hovertemplate='<b>%{text}</b><br>Surface: %{x} m²<br>Price: %{y} €<extra></extra>',
    name='Properties'
)

# Create the average curve trace
curve_trace = go.Scatter(
    x=avg_price_by_surface['Surface Group'],
    y=avg_price_by_surface['Price (€)'],
    mode='lines',
    name='Average Price Curve',
    line=dict(color='white', width=3)
)

# Build the figure manually to control layering
fig = go.Figure(data=[scatter_trace, curve_trace])

# Layout adjustments
fig.update_layout(
    template='plotly_dark',
    title='Price vs Surface in Madrid',
    xaxis_title='Surface (m²)',
    yaxis_title='Price (€)',
    height=800,
    margin=dict(l=40, r=40, t=80, b=40),
    showlegend=False
)

# Display the plot
fig.show()

# Ensure output directory exists
os.makedirs("site/assets", exist_ok=True)

# Save to HTML
fig.write_html("site/assets/madrid_scatter.html")
print("✅ Scatter plot saved to 'site/assets/madrid_scatter.html'")
```
